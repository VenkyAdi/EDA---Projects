## Introduction

The hotel industry operates in a highly competitive environment, making it imperative for hoteliers to continuously adapt and innovate to remain ahead. A critical aspect of maintaining a competitive edge is understanding customer behavior and booking trends. This can be achieved through comprehensive Exploratory Data Analysis (EDA) of hotel booking data, which provides valuable insights into customer preferences, booking patterns, and potential areas for strategic improvement. This analysis aims to reveal key aspects of customer behavior and hotel booking trends, enabling hotels to enhance their business outcomes.

## Customer Behavior Analysis

### Demographic Insights

Understanding the demographics of customers is foundational to tailoring marketing and service strategies effectively. By analyzing data on age, nationality, and travel purposes, hotels can gain a clearer picture of their customer base. For instance, identifying the predominant age groups among guests can help in customizing promotional campaigns. If the majority of customers are young adults, marketing efforts could focus on adventure activities and nightlife. Conversely, if older adults are more prevalent, promotions might highlight comfort and relaxation amenities.

Nationality analysis can reveal the most common countries of origin for guests, allowing hotels to tailor their services to meet the cultural and linguistic needs of these groups. For example, if a significant portion of guests comes from a particular country, hotels could offer services in that countrys language or include cuisine that caters to their tastes.

Understanding the primary reasons for travelbe it leisure, business, or other purposesenables hotels to customize their offerings accordingly. Leisure travelers might prioritize amenities like pools, spas, and sightseeing tours, while business travelers may value high-speed internet, meeting rooms, and quick check-in/check-out services.

### Booking Preferences

Analyzing booking preferences is crucial for optimizing operational efficiency and marketing strategies. One key metric is the lead time, which is the period between when a booking is made and the guests actual arrival. Different customer segments may exhibit varying lead times. For instance, business travelers might book closer to their travel dates, while leisure travelers may plan further in advance. Understanding these patterns helps in managing room inventory and pricing strategies more effectively, ensuring maximum occupancy and revenue.

Another important aspect is the distribution of booking channels. By examining whether bookings come through direct channels (such as the hotels website), online travel agencies (OTAs), or corporate accounts, hotels can determine which platforms are most effective. This insight allows for optimized channel management, ensuring that marketing efforts and resources are allocated to the most productive channels. For example, if OTAs generate a significant portion of bookings, investing in better positioning and promotions on these platforms could be beneficial.

## Trends and Patterns Analysis

### Seasonality and Demand Fluctuations

Seasonal trends and demand fluctuations play a significant role in hotel bookings. By plotting booking data over time, hotels can identify peak seasons and off-peak periods. Understanding these trends is crucial for effective revenue management. During peak seasons, hotels can increase room rates to capitalize on high demand. Conversely, during off-peak times, offering promotional discounts and packages can help maintain occupancy levels.

Additionally, recognizing the impact of regional events, holidays, and festivals on booking patterns is essential. For example, if a city hosts a major annual festival, hotels can anticipate a surge in bookings and adjust their strategies accordingly. This might involve increasing staff levels, enhancing specific services, or offering special packages to attract more guests.

Analyzing longer-term trends can also provide insights into how booking patterns evolve over the years. This helps in forecasting future demand and making informed decisions about expansions, renovations, or changes in services offered.

**How do booking trends differ between weekdays and weekends**

```python

# Filter the DataFrame for each hotel type
city_hotel = h1_df[h1_df['hotel'] == 'City Hotel']
resort_hotel = h1_df[h1_df['hotel'] == 'Resort Hotel']

city_weekend = city_hotel[city_hotel['is_canceled'] == 'Not Cancelled']['stays_in_weekend_nights'].sum()
city_week = city_hotel[city_hotel['is_canceled'] == 'Not Cancelled']['stays_in_week_nights'].sum()
resort_weekend = resort_hotel[resort_hotel['is_canceled'] == 'Not Cancelled']['stays_in_weekend_nights'].sum()
resort_week = resort_hotel[resort_hotel['is_canceled'] == 'Not Cancelled']['stays_in_week_nights'].sum()


data = {
    'Period': ['Weekend', 'Weekend', 'Weekday', 'Weekday'],
    'Total Stays': [city_weekend, resort_weekend, city_week, resort_week],
    'Hotel Type': ['City Hotel', 'Resort Hotel', 'City Hotel', 'Resort Hotel']
}
df = pd.DataFrame(data)

# Create the stacked bar plot
fig = px.bar(df, x='Period', y='Total Stays', color='Hotel Type',
             title='Total Stays During Weekends and Weekdays',
             labels={'Period': 'Period', 'Total Stays': 'Total Stays'},
             color_discrete_sequence=['#19AC5B', '#F69E1D'])

# Customize layout
fig.update_layout(title=dict(x=0.5, y=0.95, xanchor='center', yanchor='top'),
                  paper_bgcolor='#CAF1DE',
                  plot_bgcolor='#E1F8DC',
                  xaxis=dict(showgrid=False,gridcolor = 'black'),
                  yaxis=dict(showgrid=False,gridcolor = 'black'))

# Show the plot
fig.show()
```

**Which agent made most booking**

```python
# Group by 'hotel' and 'agent' and count the number of bookings per agent
highest_booking = h1_df.groupby(['hotel', 'agent'])['agent'].agg({'count'}).reset_index().rename(columns={'count': 'Most Bookings'}).sort_values(by='Most Bookings', ascending=False)

# Drop rows where 'agent' is 0 (assuming 0 is an invalid or placeholder value)
highest_booking.drop(highest_booking[highest_booking['agent'] == 0].index, inplace=True)

# Filter for each hotel type
city_hotel_bookings = highest_booking[highest_booking['hotel'] == 'City Hotel']
resort_hotel_bookings = highest_booking[highest_booking['hotel'] == 'Resort Hotel']

# Select the top agents for each hotel type
top_agents_limit = 5
top_city_agents = city_hotel_bookings.sort_values(by='Most Bookings', ascending=False).head(top_agents_limit)
top_resort_agents = resort_hotel_bookings.sort_values(by='Most Bookings', ascending=False).head(top_agents_limit)

# Combine the top agents data into one DataFrame
top_agents_combined = pd.concat([top_city_agents, top_resort_agents])

# Create the grouped bar chart
fig = px.bar(top_agents_combined,
             x='agent',
             y='Most Bookings',
             color='hotel',
             barmode='group',
             text='Most Bookings',
             title='Most Bookings By Top Agents',
             labels={'agent': 'Agent', 'Most Bookings': 'Number of Bookings', 'hotel': 'Hotel Type'},
             height=600, width=1200)

# Update traces for text position and bar width
fig.update_traces(textposition='inside', width=6.8)

# Update layout for title, background color, and grid lines
fig.update_layout(
    title={
        'text': 'Most Bookings By Top Agents',
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top',
    },
    plot_bgcolor='#E1F8DC',
    paper_bgcolor='#CAF1DE',
    xaxis_showgrid=False,
    yaxis_showgrid=False,
    xaxis_gridcolor='black',
    yaxis_gridcolor='black',)

# Show the plot
fig.show()
```

#### **Revenue Analysis**

**What is the average daily rate (ADR) and revenue per available room (RevPAR)**

```python
total_room_revenue = h1_df['adr'] * (h1_df['stays_in_weekend_nights'] + h1_df['stays_in_week_nights'])
total_rooms_sold = h1_df['is_cancelled_copy'].value_counts().get(0, 0)
total_available_rooms = h1_df['reserved_room_type'].value_counts().sum()
adr = int(total_room_revenue.sum() / total_rooms_sold)
revpar = int(total_room_revenue.sum() / total_available_rooms)


metrics = ['ADR', 'RevPAR']
values = [adr, revpar]

fig = go.Figure(go.Bar(
    x=metrics,
    y=values,
    text=values,
    textposition='auto',
    marker=dict(color=['blue', 'green'])
))


fig.update_layout(
    title=dict(
        text='Average Daily Rate (ADR) and Revenue per Available Room (RevPAR)',
        x=0.5,
        y=0.95,
        xanchor='center',
        yanchor='top'
    ),
    xaxis_title='Metric',
    yaxis_title='Value',
    paper_bgcolor='#CAF1DE',
    plot_bgcolor='#E1F8DC',
    xaxis=dict(showgrid=False),
    yaxis=dict(showgrid=False)
)

# Show the plot
fig.show()
```

This indicates a high level of initial commitment and satisfaction with the booking details provided at the time of reservation. About 10% of bookings, however, did see single changes made, showing that some guests needed to adjust their plans post-booking, possibly due to minor changes in their travel itinerary.

Booking channels also reveal interesting insights. Approximately 79.1% of bookings were made through travel agents or tour operators. This underscores the significant role that these intermediaries play in the travel and hospitality industry. The convenience, bundled deals, and perceived security offered by travel agents and tour operators likely drive this high percentage. It reflects guests preference for the streamlined and often more comprehensive services provided by these booking channels.
